```python
import os

import pandas as pd

os.environ['LOCAL_DEVELOPMENT'] = 'True'
os.environ['ENV'] = 'local'
os.environ['SKIP_DB_CONNECTION'] = 'True'
from ads import run_auto_ml, UserInput
```

```python
# initial data load
cpi = pd.read_csv('../data/CPALTT01USQ659N.csv', names=['date', 'cpi'], header=0)
unemp = pd.read_csv('../data/LRUN64TTUSQ156S.csv', names=['date', 'unemp'], header=0)
gov_bond_10_yr = pd.read_csv('../data/IRLTLT01USQ156N.csv', names=['date', 'gov_bond_10_yr'], header=0)
usd_yuan = pd.read_csv('../data/CCUSMA02CNM618N.csv', names=['date', 'usd_yuan'], header=0)
usd_mx_peso = pd.read_csv('../data/CCUSMA02MXM618N.csv', names=['date', 'usd_peso'], header=0)

# merge DFs
df = pd.merge(cpi, unemp, on='date')
df = pd.merge(df, gov_bond_10_yr, on='date')
df = pd.merge(df, usd_yuan, on='date')
df = pd.merge(df, usd_mx_peso, on='date')
print(df.shape)
df.head(2)
```

```python

user_input = { 
    'problemType':'regression',
    'numericColumns':[ 'unemp', 'gov_bond_10_yr', 'usd_yuan', 'usd_peso'],
    'targetColumn': 'cpi',
    #'categoricalColumns':[],
    'workflow':'autoMl',
    'test_size':0.20,
    'shuffle':False,
    'scaler':'standard',
    'n_iter':50,
    'k_fold':5,
    'imputer_strategy':'mean',
    'n_param_samples':5,
    'feature_selection':True
 }
```

```python


result = run_auto_ml(df, user_input)
result
```
